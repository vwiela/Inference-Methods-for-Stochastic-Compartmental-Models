```python
import cnfcomp
from cnfcomp import input_result_plot
from cnfcomp import plot_results
from cnfcomp import plot_histograms
import numpy as np
import pandas as pd
```

```python
sir_prior_dict = cnfcomp.sir_prior_wrap()

config_sir_1 = cnfcomp.load_data_dense('./data/sir_dense_1.csv', 'sir_dense_1', T1=100)
config_sir_2 = cnfcomp.load_data_dense('./data/sir_dense_2.csv', 'sir_dense_2', T1=100)
```

```python
folder_cnf = "./output/CNF_Experiments/" 
folder_pf = "./output/PNF_Experiments/"

postsamples_nf_sir_dense_1 = np.array(
    pd.read_csv(folder_cnf + "sir_dense_1/posterior_samples.csv", sep=",")
)
postsamples_nf_sir_dense_2 = np.array(
    pd.read_csv(folder_cnf + "sir_dense_2/posterior_samples.csv", sep=",")
)

postsamples_pf_sir_dense_1 = np.array(
    pd.read_csv(folder_pf + "sir_dense_1/posterior_samples.csv", sep=",")
)
postsamples_pf_sir_dense_2 = np.array(
    pd.read_csv(folder_pf + "sir_dense_1/posterior_samples.csv", sep=",")
)

true_params_1 = np.array([0.1, 20])
true_params_2 = np.array([0.22, 5])
```

```python
input_sir_dense_1 = input_result_plot(
    postsamples_nf_sir_dense_1,
    postsamples_pf_sir_dense_1,
    config_sir_1,
    "normal",
    ["$\\beta$", "$\\gamma^{-1}$"],
    "sir_dense_1",
    true_params_1,
)

f = plot_results(
    input_sir_dense_1, lower_b=np.array([0.095, 17]), upper_b=np.array([0.11, 21])
)
f.savefig("./figures/comparison_sir_dense_1.pdf")
print("Sir Dense 1 Done!")

input_sir_dense_2 = input_result_plot(
    postsamples_nf_sir_dense_2,
    postsamples_pf_sir_dense_2,
    config_sir_2,
    "normal",
    ["$\\beta$", "$\\gamma^{-1}$"],
    "sir_dense_2",
    true_params_2,
)

f = plot_results(input_sir_dense_2, lower_b=np.array([0.2, 4]), upper_b=np.array([0.28, 6]))
f.savefig("./figures/comparison_sir_dense_2.pdf")
print("Sir Dense 2 Done!")
```
