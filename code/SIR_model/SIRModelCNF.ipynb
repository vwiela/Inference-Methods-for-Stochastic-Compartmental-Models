```python
import cnfcomp
import numpy as np
import pandas as pd
import os 
os.chdir("../../")
basefolder = "./code/"
resultsfolder = "./output/"
figuresfolder = "./figures/"
```

```python
sir_prior_dict = cnfcomp.sir_prior_wrap()

config_sir_1 = cnfcomp.load_data_dense('./data/sir_dense_normal_and_binomial_1.csv', 'sir_dense_1', T1=100)
config_sir_2 = cnfcomp.load_data_dense('./data/sir_dense_normal_and_binomial_2.csv', 'sir_dense_2', T1=100)
```

```python
cnfcomp.train_validate_infer(config_sir_1, sir_prior_dict, resultsfolder + "SimulationRuns/sir_dense_1_paper/", "sir_dense_1_paper", model = "sir", summary_dim = 6)
cnfcomp.train_validate_infer(config_sir_2, sir_prior_dict, resultsfolder + "SimulationRuns/sir_dense_2_paper/", "sir_dense_1_paper", model = "sir", summary_dim = 6)
```

```python
from cnfcomp import input_result_plot
from cnfcomp import plot_results

postsamples_nf_sir_full_normal_dense_1 = np.array(
    pd.read_csv(resultsfolder + "Bayesflow_100epoch/" + "posterior_sir_dense_1_paper.csv", sep=",")
)
postsamples_nf_sir_full_normal_dense_2 = np.array(
    pd.read_csv(resultsfolder + "Bayesflow_100epoch/" + "posterior_sir_dense_2_paper.csv", sep=",")
)

postsamples_pf_sir_full_normal_dense_1 = np.array(
    pd.read_csv(
        resultsfolder + "ParticleFilter/" + "posterior_sir_full_normal_dense_1.csv", sep=","
    )
)
postsamples_pf_sir_full_normal_dense_2 = np.array(
    pd.read_csv(
        resultsfolder + "ParticleFilter/" + "posterior_sir_full_normal_dense_2.csv", sep=","
    )
)

true_params_1 = np.array([0.1, 20])
true_params_2 = np.array([0.22, 5])

input_sir_full_dense_1 = input_result_plot(
    postsamples_nf_sir_full_normal_dense_1,
    postsamples_pf_sir_full_normal_dense_1,
    config_sir_1,
    "normal",
    ["$\\beta$", "$\\gamma^{-1}$"],
    "sir_dense_1",
    true_params_1,
)

f = plot_results(
    input_sir_full_dense_1, lower_b=np.array([0.095, 17]), upper_b=np.array([0.11, 21])
)
f.savefig(figuresfolder + "trajectories_sir_full_dense_1.pdf")
print("Sir Dense 1 Done!")

input_sir_full_dense_2 = input_result_plot(
    postsamples_nf_sir_full_normal_dense_2,
    postsamples_pf_sir_full_normal_dense_2,
    config_sir_2,
    "normal",
    ["$\\beta$", "$\\gamma^{-1}$"],
    "sir_dense_2",
    true_params_2,
)

f = plot_results(input_sir_full_dense_2, lower_b=np.array([0.2, 4]), upper_b=np.array([0.28, 6]))
f.savefig(figuresfolder + "trajectories_sir_full_dense_2.pdf")
print("Sir Dense 2 Done!")
```
