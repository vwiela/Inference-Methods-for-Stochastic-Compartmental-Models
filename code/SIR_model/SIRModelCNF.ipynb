```python
import cnfcomp
import numpy as np
import pandas as pd
import os 
os.chdir("../../")
basefolder = "./code/"
resultsfolder = "./output/"
figuresfolder = "./figures/"
```

```python
sir_prior_dict = cnfcomp.sir_prior_wrap()

config_sir_1 = cnfcomp.load_data_dense('./data/sir_dense_normal_and_binomial_1.csv', 'sir_dense_1', T1=100)
config_sir_2 = cnfcomp.load_data_dense('./data/sir_dense_normal_and_binomial_2.csv', 'sir_dense_2', T1=100)
```

```python
cnfcomp.train_validate_infer(config_sir_1, sir_prior_dict, resultsfolder + "SimulationRuns/sir_dense_1_paper/", "sir_dense_1_paper", observationtype = "normal", model = "sir", summary_dim = 4, epochs = 100)
cnfcomp.train_validate_infer(config_sir_2, sir_prior_dict, resultsfolder + "SimulationRuns/sir_dense_2_paper/", "sir_dense_1_paper", observationtype = "normal", model = "sir", summary_dim = 4, epochs = 100)
```

```python
from cnfcomp import plot_input
from cnfcomp import plot_posteriors_2d_sir

n = 10000

postsamples_nf_sir_full_normal_dense_1 =  np.array(pd.read_csv(resultsfolder + "Bayesflow_100epoch/" + "posterior_sir_full_normal_dense_1_s4.csv", sep=","))[0:n,]
postsamples_nf_sir_full_normal_dense_2 =  np.array(pd.read_csv(resultsfolder + "Bayesflow_100epoch/" + "posterior_sir_full_normal_dense_2.csv", sep=","))[0:n,]

postsamples_pf_sir_full_normal_dense_1 =  np.array(pd.read_csv(resultsfolder + "ParticleFilter/" + "posterior_sir_full_normal_dense_1.csv", sep=","))[0:n,]
postsamples_pf_sir_full_normal_dense_2 =  np.array(pd.read_csv(resultsfolder + "ParticleFilter/" + "posterior_sir_full_normal_dense_2.csv", sep=","))[0:n,]

true_params_1 = np.array([0.1, 20])
true_params_2 = np.array([0.22, 5])

input_sir_full_dense_1 = plot_input(postsamples_nf_sir_full_normal_dense_1, postsamples_pf_sir_full_normal_dense_1, config_sir_1, "normal", 
                                         [ "$\\beta$", "$\\gamma^{-1}$"], "sir_dense_1", true_params_1)

f = plot_posteriors_2d_sir(input_sir_full_dense_1)
f.savefig(figuresfolder + "trajectories_sir_full_dense_1.svg", format="svg", transparent=True)
print("Sir Dense 1 Done!")

input_sir_full_dense_2 = plot_input(postsamples_nf_sir_full_normal_dense_2, postsamples_pf_sir_full_normal_dense_2, config_sir_2, "normal", 
                                         [ "$\\beta$", "$\\gamma^{-1}$"], "sir_dense_2", true_params_2)

f = plot_posteriors_2d_sir(input_sir_full_dense_2)
f.savefig(figuresfolder + "trajectories_sir_full_dense_2.svg", format="svg", transparent=True)
print("Sir Dense 2 Done!")
```
