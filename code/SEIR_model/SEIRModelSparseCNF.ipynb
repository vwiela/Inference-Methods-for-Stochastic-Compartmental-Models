```python
seir2v_prior_full_dict = cnfcomp.seir2v_prior_full_wrap()

config_seir2v_sparse_1_1_1 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_1_1.csv', 'sparse_1_1_1')
config_seir2v_sparse_1_1_2 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_1_2.csv', 'sparse_1_1_2')
config_seir2v_sparse_1_1_3 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_1_3.csv', 'sparse_1_1_3')
config_seir2v_sparse_1_2_1 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_1_1.csv', 'sparse_1_2_1')
config_seir2v_sparse_1_2_2 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_2_2.csv', 'sparse_1_2_2')
config_seir2v_sparse_1_2_3 = cnfcomp.load_data_sparse('./data/seir2v_synth_sparse_binomial_and_normal_1_2_3.csv', 'sparse_1_2_3')

config_seir2v_sparse_simon = cnfcomp.load_data_sparse('./data/seir2v_ethiopia_simon.csv', 'simon')
```
